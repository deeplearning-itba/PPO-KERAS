```python
from matplotlib import pyplot as plt
from IPython.display import clear_output
import numpy as np
```

```python
from gym.envs.box2d.car_dynamics import Car
from gym.envs.box2d import CarRacing
```

```python
env = gym.make('CarRacing-v0')
```

```python
env.action_space
```

```python
env.action_space.high, env.action_space.low
```

```python
def get_random_action():
    action = np.random.random(3)
    action[0] = action[0]*2-1
    return action
```

```python
np.random.randint(3)
```

```python
def get_dicrete_random_action():
    action = np.random.randint(2, size=3)
    action[0] = action[0]*2-1
    return action
```

```python
get_dicrete_random_action()
```

```python
done = False
total_reward = 0
rewards = []
total_rewards = []
episode = 0
while episode<100:
    obs, reward, done, info = env.step([0,  0,  0])
    rewards.append(reward)
    total_reward += reward
    total_rewards.append(total_reward)
    clear_output(wait=True)
    plt.imshow(obs)
    plt.show()
    episode+=1
    print(f'\r{reward}, {done}, {info}, {total_reward}, {episode}', end='')
```

```python
n=100
plt.plot(rewards[:n])
plt.plot(total_rewards[:n])
```

```python
obs = env.reset()
done = False
total_reward = 0
rewards = []
total_rewards = []
episode = 0
while not done:
    obs, reward, done, info = env.step(get_dicrete_random_action())
    rewards.append(reward)
    total_reward += reward
    total_rewards.append(total_reward)
    clear_output(wait=True)
    plt.imshow(obs)
    plt.show()
    episode+=1
    print(f'\r{reward}, {done}, {info}, {total_reward}, {episode}', end='')
#     print(obs_step[1], obs_step[2], obs_step[3])
```
